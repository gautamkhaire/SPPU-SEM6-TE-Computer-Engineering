### Categorical to Numerical data

```python
# Categorical to Numerical
# Label Encoding Technique -> Index numbering
# One-Hot Encoding -> Male, Female, Others (No. catergory is no. of columns).

# Converting datatypes of location code and location columns to category datatype
cat_features = ['location code','location']
data[cat_features] = data[cat_features].astype('category')
```

```python
cat_columns = data.select_dtypes(['category']).columns
print(cat_columns)
data[cat_columns] = data[cat_columns].apply(lambda x: x.cat.codes)
```
